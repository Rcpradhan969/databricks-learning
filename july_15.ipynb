```python

df=spark.read.format("csv")\
    .option("header","true")\
        .option("inferSchema","true")\
            .load("dbfs:/Volumes/workspace/practice/my_volume_1/automobile_dataset.csv")
```

```python
df.write.mode("overwrite").saveAsTable("cars_bronze")
```

```python
silver_df = df.dropDuplicates()
```

```python
df = df.dropna(
    subset=["Selling_Price"]
)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
silver_df = df.filter(
    col("Selling_Price") > 0
)
```

```python
silver_df = silver_df.filter(
    col("Mileage") > 0
)
```

```python
silver_df = silver_df.fillna({

"Color":"Unknown",
"Location":"Unknown"

})
```

```python
silver_df = silver_df.filter(
    (col("Year") >=1990) &
    (col("Year")<=2026)
)
```

```python
silver_df = silver_df.fillna(
    {
        "Fuel_Efficiency":0
    }
)
```

```python
silver_df = silver_df.filter(
    col("Owners") <=10
)
```

```python
silver_df.write.mode("overwrite").saveAsTable("cars_silver")
```

```python
gold_price = silver_df.groupBy("Make")\
.agg(
avg("Selling_Price").alias(
"Average_Selling_Price"
)
)
```

```python
gold_price.write.mode("overwrite").saveAsTable(
"gold_avg_selling_price"
)
```

```python
gold_location = silver_df.groupBy("Location")\
.count()
```

```python
gold_location.write.mode("overwrite").saveAsTable(
"gold_location_summary"
)
```

```python
gold_fuel = silver_df.groupBy(
"Fuel_Type"
).agg(
avg("Selling_Price").alias(
"Avg_Selling_Price"
)
)
```

```python
gold_top10 = silver_df.orderBy(
col("Selling_Price").desc()
).limit(10)
```

```python
gold_year = silver_df.groupBy(
"Year"
).count()
```
